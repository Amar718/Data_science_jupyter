````python
Q1. Probability density function (PDF) is a function that describes the relative likelihood for this random variable to take on a given value. It is used to specify the probability of the random variable falling within a particular range of values, as opposed to taking on any one value. ¹ 

Q2. There are many types of probability distributions. Some of them are:
- Uniform Distribution
- Normal Distribution
- Binomial Distribution
- Poisson Distribution
- Exponential Distribution
- Gamma Distribution
- Beta Distribution
- Lognormal Distribution ²

Q3. Here is an example Python function that calculates the probability density function of a normal distribution with given mean and standard deviation at a given point:

```python
import math

def normal_pdf(x, mu=0, sigma=1):
    sqrt_two_pi = math.sqrt(2 * math.pi)
    return (math.exp(-(x-mu)**2 / 2 / sigma**2) / (sqrt_two_pi * sigma))
```

Q4. The properties of Binomial distribution are:
- The experiment consists of n repeated trials.
- Each trial can result in just two possible outcomes.
- The probability of success is constant throughout the experiment.
- The trials are independent.

Two examples of events where binomial distribution can be applied are:
- Tossing a coin.
- Rolling a dice.

Q5. Here is an example Python code that generates a random sample of size 1000 from a binomial distribution with probability of success 0.4 and plots a histogram of the results using matplotlib:

```python
import numpy as np
import matplotlib.pyplot as plt

n, p = 1000, 0.4  # number of trials, probability of each trial

# Generate random samples from binomial distribution
samples = np.random.binomial(n, p, size=1000)

# Plot histogram of samples
plt.hist(samples, bins=30)
plt.show()
```

Q6. Here is an example Python function that calculates the cumulative distribution function (CDF) of a Poisson distribution with given mean at a given point:

```python
import math

def poisson_cdf(k, lamb):
    cdf = 0
    for i in range(k+1):
        cdf += math.exp(-lamb) * lamb**i / math.factorial(i)
    return cdf
```

Q7. The main difference between Binomial and Poisson distributions is that Binomial distribution is used when there are a fixed number of trials and each trial has only two possible outcomes whereas Poisson distribution is used when there are an infinite number of trials and each trial has only one possible outcome.

Q8. Here is an example Python code that generates a random sample of size 1000 from a Poisson distribution with mean 5 and calculates the sample mean and variance:

```python
import numpy as np

# Generate random samples from Poisson distribution
samples = np.random.poisson(5, size=1000)

# Calculate sample mean and variance
mean = np.mean(samples)
variance = np.var(samples)

print("Sample mean:", mean)
print("Sample variance:", variance)
```

Q9. In Binomial distribution, the mean and variance are related by the formula: mean = n * p and variance = n * p * (1 - p), where n is the number of trials and p is the probability of success in each trial. In Poisson distribution, the mean and variance are equal.

Q10. In normal distribution with respect to mean position, the least frequent data appears at two standard deviations below or above the mean position ³.

````
